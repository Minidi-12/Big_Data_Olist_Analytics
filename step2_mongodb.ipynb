```python
print('=== WHY NOSQL FOR E-COMMERCE DATA ===')
print()
print('1. SCHEMA FLEXIBILITY')
print('   Each order document can have different fields.')
print('   Example: some orders have installments, others do not.')
print('   In SQL this needs nullable columns or extra tables.')
print('   In MongoDB each document is independent JSON.')
print()
print('2. HORIZONTAL SCALABILITY')
print('   MongoDB Atlas shards data across multiple servers.')
print('   As order volume grows, add more servers automatically.')
print('   SQL databases scale vertically (bigger single server).')
print()
print('3. NATIVE JSON FORMAT')
print('   E-commerce APIs already produce JSON.')
print('   No transformation needed before storage.')
```
